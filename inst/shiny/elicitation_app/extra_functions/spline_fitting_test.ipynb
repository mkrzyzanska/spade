```r
# Load necessary packages
library(splines2)
set.seed(123)  # For reproducibility

# Generate synthetic data
x <- seq(-3, 3, length.out = 100)  
y <- sin(2 * x) + rnorm(100, sd = 0.2)  # Noisy sine wave data

# Define spline parameters
degree <- 3  # Cubic B-spline
knots <- seq(-2, 2, length.out = 5)  # Interior knots

# Generate B-spline basis
bs_basis <- bSpline(x, knots = knots, degree = degree, intercept = TRUE)

# Apply exponential decay
lambda <- 1.5  # Decay rate
decay_function <- exp(-lambda * abs(x))  # Exponential decay
bs_basis_decay <- bs_basis * decay_function  # Apply decay to basis

# Fit linear model using B-spline basis
model <- lm(y ~ bs_basis_decay)

# Make predictions
y_pred <- predict(model)

# Plot results
plot(x, y, pch = 16, col = "gray", main = "Fitting B-Spline with Exponential Decay")
lines(x, y_pred, col = "blue", lwd = 2)
legend("topright", legend = c("Data", "Fitted Spline"), col = c("gray", "blue"), lty = c(NA, 1), pch = c(16, NA))
```

```r
# Load necessary package
library(splines2)

# Set seed for reproducibility
set.seed(123)

# Generate data similar to a normal distribution
x <- seq(-3, 3, length.out = 100)  # X values
y <- dnorm(x, mean = 0, sd = 1) + rnorm(100, sd = 0.02)  # Noisy normal distribution

# Define B-spline parameters
degree <- 3  # Cubic B-spline
knots <- seq(-2, 2, length.out = 5)  # Interior knots

# Generate B-spline basis
bs_basis <- bSpline(x, knots = knots, degree = degree, intercept = TRUE)

# Apply exponential decay
lambda <- 1.5  # Decay rate
decay_function <- exp(-lambda * abs(x))  # Exponential decay function
bs_basis_decay <- bs_basis * decay_function  # Apply decay

# Fit a model using the modified B-spline basis
model <- lm(y ~ bs_basis_decay)

# Predict values using the model
y_pred <- predict(model)

# Plot the original data and the fitted spline
plot(x, y, pch = 16, col = "gray", main = "Fitting B-Spline to Normal-Like Data")
lines(x, y_pred, col = "blue", lwd = 2)
legend("topright", legend = c("Data", "Fitted Spline"), col = c("gray", "blue"), 
       lty = c(NA, 1), pch = c(16, NA))
```
